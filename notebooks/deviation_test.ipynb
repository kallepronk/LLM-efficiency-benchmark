```python
# With encoding
model: str = "EleutherAI/pythia-70m"
dataset: Dataset = Dataset(name="Rowan/hellaswag", split="train", column="ctx")

# Configure runs
warmup: Run = Warmup(model=model, dataset=dataset, passes=100)
runs = [warmup]
for i in range(50):
    runs.append(PassesRun(model=model, dataset=dataset, passes=20))

# Configure classes
benchmark = Benchmark(name="deviation_20p", runs=runs)

benchmark.run()

print(benchmark.collect_results())
```
